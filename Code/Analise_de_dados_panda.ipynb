#7.Qual a moeda em que mais salários foram pagos (salary_currency)?

```python
salariosemmoedas = df[['salary','salary_currency']].groupby('salary_currency').sum().sort_values(by='salary', ascending = False)
salariosemmoedas
```

#8 -Quais as médias e medianas dos salários pagos, em dólar, para cada cargo?

```python
df[['job_title','salary_in_usd']].groupby('job_title').agg([np.mean,np.median])
```

#Quais os valores totais dos salários pagos( as somas de todos os salários  pagos  por  cada  país),  em  dólar,  para  cada  país  contratante (company_location)?

```python
salariospagos = df[['salary_in_usd','company_location']].groupby('company_location').sum().sort_values(by='salary_in_usd', ascending = False)
salariospagos
```

#10.Qual o país contratante que possui o maior valor total pago, em dólar, em salários?

```python
salariospagos = df[['salary_in_usd','company_location']].groupby('company_location').sum().sort_values(by='salary_in_usd', ascending = False).max()
salariospagos
```
